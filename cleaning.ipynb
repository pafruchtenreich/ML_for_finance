```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Loading and cleaning data

```python
fed_rates = pd.read_csv("fed_rates.csv")
gold_prices = pd.read_csv("gold_historical_data.csv")
```

## Gold prices

```python
gold_prices = gold_prices.drop([0,1]).rename(columns={"Price":"Date"})
```

```python
gold_prices["Date"] = pd.to_datetime(gold_prices["Date"])
gold_prices["Year"] = gold_prices["Date"].dt.year
gold_prices["Month"] = gold_prices["Date"].dt.month
gold_prices["Day"] = gold_prices["Date"].dt.day
```

```python
gold_prices.drop(columns=["Date"], inplace=True)
```

```python
cols = gold_prices.columns.to_list()
new_cols = cols[-3:] + cols[:-3]
gold_prices = gold_prices[new_cols]
```
